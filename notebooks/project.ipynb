```python
import re

file = open('../data/baby1990.html')

year = []
pattern = r"\b(?:19|20)\d{2}\b"

for line in file:
    line = line.strip()# Elimina renglones en blanco
    result = re.findall(pattern, line)
    if result:
        year.extend(result)
        break
    #print(line)

print(year, len(year))
```

```python
file = open('../data/baby1990.html')
pattern = r"<tr[^>]*>\s*<td>(\d+)</td>\s*<td>([A-Za-z'-]+)</td>\s*<td>([A-Za-z'-]+)</td>"
pairs = []
for line in file:
    line = line.strip()
    result = re.findall(pattern, line)
    if result:
        for rank, girl, boy in result:
            pairs.append((girl, rank))
            pairs.append((boy, rank))

file.close()

print(pairs)
print('Total de tuplas:', len(pairs))
```

```python
name_rank = {}

for name, rank in pairs:

    if name not in name_rank or int(rank) < int(name_rank[name]):
        name_rank[name] = rank

for name, rank in name_rank.items():
    print(name, "-", rank)
print("Total de nombres únicos:", len(name_rank))
```

```python
names_sorted = sorted(name_rank.keys())

print(names_sorted)
print("Total de nombres:", len(names_sorted))
```

```python
final_output = [year]

for name in sorted(name_rank.keys()):
    final_output.append(f"{name} - {name_rank[name]}")

print(final_output[:20])
print("Total de elementos:", len(final_output))
```
